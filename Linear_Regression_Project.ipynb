```python
import pandas as pd
import plotly.express as plot
import numpy as np
import statistics as st
```

```python
df = pd.read_csv('score.csv')
df.head()
```

```python
#Data seperation
toefl = df['TOEFL Score']
chance_of_admit = df['Chance of Admit ']
```

```python
#Scatterplot
plotter_1 = plot.scatter(x=toefl, y=chance_of_admit, color=chance_of_admit)
plotter_1.show()
```

```python
#Personal mathematical method of finding a good fit for a linear regression
#Initialisation
m = 0
c = 0

#Sorting Data (Dependant)
data_1 = toefl.tolist()
data_2 = chance_of_admit.tolist()

def sorter_backend(data1, data2):
  sorter = [[], []]
  for i in range(len(data1)):
    if min(data1) == data1[i]:
      sorter[0].append(data1[i])
      sorter[1].append(data2[i])
      sorter.append(i)
      break
    else:
      pass
  return sorter

def sorter(data1, data2):
  sorted_data = [[], []]

  while len(data1) != 0:
    d1, d2, ind = sorter_backend(data1, data2)
    data1.pop(ind)
    data2.pop(ind)
    sorted_data[0].append(d1[0])
    sorted_data[1].append(d2[0])
  return sorted_data[0], sorted_data[1]

#Best M and C predictor
def fit(x, y, manual):
  global m_mean
  global c_mean
  import statistics as st
  m_list = []
  c_list = []
  for i in range(len(x)):
    if i < len(x)-1:
      x0 = x[i]
      y0 = y[i]
      x1 = x[i+1]
      y1 = y[i+1]

      try:
        m_temp = (y1 - y0)/(x1 - x0)
        m_list.append(m_temp)
      except:
        m_temp = (y1 - y0)/manual
        m_list.append(m_temp)

    else:
      break

  m_mean = st.mean(m_list)

  for i in range(len(x)):
    x0 = x[i]
    y0 = y[i]

    c_temp = y0 - (m_mean*x0)
    c_list.append(c_temp)

  c_mean = st.mean(c_list)

  slope_co = []

  for x0 in x:
    y_temp = (m_mean*x0) + c_mean
    slope_co.append(y_temp)

  return slope_co

sort_d1, sort_d2 = sorter(data_1, data_2)
slope_data = fit(sort_d1, sort_d2, 0.12)


plottertest = plot.scatter(x = toefl, y = chance_of_admit)
plottertest.update_layout(shapes=[dict(type='line',y0=min(slope_data), y1=max(slope_data), x0=min(toefl), x1=max(toefl))])
plottertest.show()
```

```python
np_toefl = np.array(toefl)
np_coa = np.array(chance_of_admit)
m, c = np.polyfit(np_toefl, np_coa, 1)
np_y_array = []
for x in np_toefl:
  any = (m*x)+c
  np_y_array.append(any)

plotter4 = plot.scatter(x=toefl, y=chance_of_admit)
plotter4.update_layout(shapes=[dict(type='line',y0=min(np_y_array), y1=max(np_y_array), x0=min(np_toefl), x1=max(np_toefl))])
plotter4.show()
```

```python
#Comparison of personal outcome and numpy

#Mean difference in outcomes
diff_list = []
for i in range(len(slope_data)):
  temp_diff = abs(np_y_array[i] - slope_data[i])
  diff_list.append(temp_diff)

mean_diff = st.mean(diff_list)
print('Mean difference:', mean_diff)
print('\n')
#Difference in predictions (small and large)
def predict(predictor):
  personal_prediction = (m_mean * predictor) + c_mean
  numpy_prediction = (m * predictor) + c
  print(f"Personal prediction for {predictor} is {personal_prediction}")
  print(f"Numpy prediction for {predictor} is {numpy_prediction}")

#Small (X values are closer to actual data)
print("Small prediction:")
predict(140)
print('\n')
#Large (X values are further from actual data)
print("Large prediction:")
predict(400)
```
